#### Exercici 1: Calculadora de l'índex de massa corporal
Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en:

Índice de masa corporal IMC que es y como se calcula.
La funció ha de classificar el resultat en les seves respectives categories


Categorías de IMC:

    Bajo peso: IMC < 18.5.
    Peso normal: IMC entre 18.5 y 24.9.
    Sobrepeso: IMC entre 25.0 y 29.9.
    Obesidad: IMC 30.0 o más.

````python
#Exercici 1: Calculadora de l'índex de massa corporal

def obtener_dato_numerico(params):
    """Función para solicitar datos numéricos al usuario \n
        Recibe inputs y valida por intervalos definidos
            inputs: params(dict) - contiene: mensaje(str), min_val(float) y max_val(float)
            output: valor(float) - valor numérico ingresado por el usuario
     """
    while True:
        try:
            valor = float(input(params['mensaje']))
            # Verificar si se deben aplicar ambos límites
            if 'min_val' in params and 'max_val' in params:
                if not (params['min_val'] <= valor <= params['max_val']):
                    print(f"Error: Fuera de rango ({params['min_val']}-{params['max_val']})")
                    continue
            else:
                # Verificar mínimo individual
                if 'min_val' in params and valor < params['min_val']:
                    print(f"Error: Debe ser al menos {params['min_val']}")
                    continue
                # Verificar máximo individual
                if 'max_val' in params and valor > params['max_val']:
                    print(f"Error: No debe exceder {params['max_val']}")
                    continue
            return valor
        except ValueError:
            print("Debe ser un número válido")


def calcular_imc(peso, altura):
    """Función de cálculo de IMC \n
            inputs: peso(float), altura(float) \n
            output: IMC(float)"""
    return round(peso / (altura**2),2)


def categoria_imc(imc):
    """Función para determinar la categoría del IMC \n
        input: imc(float) \n
        output: categoria(str)"""
    if imc < 18.5:
        return "Bajo peso"
    elif imc < 25:
        return "Peso normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"


def imprimir_resultados(*args):
    """Función para imprimir resultados, asociados a un texto. Par string-variable"""
    if len(args) % 2 != 0:
        raise ValueError("Debe ser pares categoria-valor")
    for i in range(0, len(args), 2):
        categoria = args[i]
        valor = args[i+1]
        print(f"{categoria}: {valor}")


def main_imc():
    """Función principal de cálculo de IMC"""
    params_peso = {
        'mensaje': 'Ingrese su peso en kg: ',
        'min_val': 0,
        'max_val': 200
        }
    
    params_altura = {
        'mensaje': 'Ingrese su altura en metros: ',
        'min_val': 0,
        'max_val': 2.5
    }
    
    peso = obtener_dato_numerico(params_peso)
    altura = obtener_dato_numerico(params_altura)
    resultado = calcular_imc(peso, altura)
    categoria = categoria_imc(resultado)
    imprimir_resultados("Su peso ingresado es", peso, 
                        "Su altura ingresada es", altura,  
                        "IMC calculado", resultado,
                        "Categoría", categoria)
````

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions.

Exercici 2: Convertidor de temperatures - Celsius (°C), Fahrenheit (°F) i Kelvin (K)

Función de cálculo - conversión de temperaturas

Este diccionario me entrega una tupla (C,F,K).

- Comentarios de codigos:

````python
def convertir_temperatura(temp, unidad):
    """Función que calcula conversión de temperatura a diferentes unidades \n
            inputs: temperatura(float), unidad(str)
            output(dict) - contiene pares unidad:temp"""
    dict1 = {'C': (round(temp,2), round((temp * 9/5) + 32,2), round(temp + 273.15,2)),
             'F': (round((temp - 32) * 5/9,2), round(temp), round((temp - 32) * 5/9 + 273.15,2)),
             'K': (round(temp - 273.15,2), round((temp - 273.15) * 9/5 + 32,2), round(temp,2))
            }
    return dict1[unidad]

def obtener_dato_categorico(mensaje, opciones_validas):
    """Función para solicitar datos al usuario - Inputs y validación (No Case Sensitive)
    entradas: (str) Texto para solicitar input categórico
    - mensaje"""
    while True:
        entrada = input(mensaje).strip().lower()
        if entrada in [opc.lower() for opc in opciones_validas]:
            return entrada.capitalize()
        print(f"Error: Opción no válida. Use: {', '.join(opciones_validas)}")


def unidad_to_param(unidad):
    """Función que transforma unidad de temperatura, en diccionario con parámetros de unidad y min_val"""
    params = {
        'C': -273.15,
        'F': -459.67,
        'K': 0
        }
    return {'mensaje': 'Ingrese la temperatura a convertir:',
            'min_val': params[unidad]}

def main_temp():
    """Función principal - conversor de temperaturas"""
    unidades = ['C','F','K']
    seleccion_unidad = obtener_dato_categorico(
    f"Ingrese la unidad de la temperatura a convertir del siguiente listado ({', '.join(unidades)}): ",
    unidades)
    temperatura = obtener_dato_numerico(unidad_to_param(seleccion_unidad))
    celsius, fahrenheit, kelvin = convertir_temperatura(temperatura, seleccion_unidad)
    imprimir_resultados("Su temperatura en °Celsius es", celsius,
                        "Su temperatura en °Fahrenheit es", fahrenheit,
                        "Su temperatura en Kelvin es", kelvin, )

````

#### Exercici 3: Comptador de paraules d'un text

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

!”#$%&'()*+,-./:;<=>?@[\]^_`{|}~

Algunos símbolos (como *, ?, [, \) tienen significados especiales en regex. Usamos re.escape() para "neutralizarlos":

    #sorted(iterable, key=key, reverse=reverse) 
    #.items() Convierte el diccionario contador en una lista de tuplas (palabra, frecuencia).

````python
import re
import string
from collections import Counter #Conteo de elementos iterables - crea un diccionario

def atrapa_palabras(mensaje):
    """Funcion que recibe texto y lo convierte en un listado de palabras separadas y limpias"""
    entrada = input(mensaje)
    texto_limpio = re.sub( r'[{}]'.format(re.escape(string.punctuation + '¡¿')),
                           '', entrada).lower().split()

    return texto_limpio


def contador_palabras(texto_lista):
    """Función para contar la frecuencia de palabras en una lista
    input: lista str []
    output: diccionario {str:frecuencia}"""
    dict_contador = Counter(texto_lista)
    return dict_contador

def ordenar_resultados(dict_contador):
    palabras_ordenadas = sorted(dict_contador.items(), key=lambda x: (-x[1], x[0]))
    return palabras_ordenadas

def imprimir_diccionario(palabras_ordenadas):
    print("\nFrecuencia de palabras:")
    for palabra, frecuencia in palabras_ordenadas:
        print(f"'{palabra}': {frecuencia}")



def main_contador():
    texto_lista = atrapa_palabras("Ingrese el texto completo del cual desea contar la frecuencia de palabras")
    dict_contador = contador_palabras(texto_lista)
    palabras_ordenadas = ordenar_resultados(dict_contador)
    imprimir_diccionario(palabras_ordenadas)
````

#### Exercici 4: Diccionari invers

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

reverse_dictionary ({'a': 1, 'b': 2, 'c': 3})
{1: 'a', 2: 'b', 3: 'c'}
reverse_dictionary({'x': 'apple', 'y': 'banana', 'z': 'banana'})
Error: multiple keys for one value


#{clave: valor for variable in coleccion if condicion}
#items()	Returns a list containing a tuple for each key value pair (LISTA DE TUPLAS)

````python
dict4 = {'a': 1,
         'b': 2,
         'c': 3,
         'd': 4,
         'e': 5,
         'f': 6,
         'g': 7,
         'h': 10,
         'i': 14
         }

dict5 = {'a': 1,
         'b': 2,
         'c': 3,
         'd': 4,
         'e': 4,
         'f': 4,
         'g': 7,
         'h': 10,
         'i': 14
         }

def tiene_duplicados(diccionario):
    """Función para ver si hay duplicados en los valores
    Obtiene valores de diccionario y entrega una lista
    Verifica longuitud de lista vs. longuitud de lista convertida a conjunto(set)
    set elimina los duplicados de una lista"""
    valores = diccionario.values()
    return len(valores) != len(set(valores))

def invertir_dic(dict):
    return {v: k for k, v in dict.items()}


def main_diccionario_inverso(dict):
    if tiene_duplicados(dict) == True:
        print('Error: multiple keys for one value')
    else:
        print(invertir_dic(dict))
````

#### Exercici 1: Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.


reverse_dictionary_plus({'x': 'apple', 'y': 'banana', 'z': 'banana'})

{'apple': 'x', 'banana': ['y', 'z']}

#defaultdict(list) crea un diccionario donde cada nueva clave se inicializa automáticamente con una lista vacía [].

Para cada entrada en invertido:
Clave: v (valor del diccionario original).
Valor: k[0] si la lista k tiene un solo elemento (para evitar listas innecesarias).
       k (la lista completa) si hay múltiples elementos.


````python
from collections import defaultdict

def invertir_dic_lista(dicc):
    """Función que invierte diccionario y aceptar valores repetidos de valores en diccionario de entrada"""
    invertido = defaultdict(list) #Diccionario vacio
    for k, v in dicc.items():
        invertido[v].append(k)  #invertido[v] -> Si v no existe como clave, la crea como clave
                                #Si existe, accede a la lista existente
                                #Con .append(k), agregamos la clave original k a la lista de v.

    return {                            #Transformo en diccionario normal 
        v: k[0] if len(k) == 1 else k   #Si hay solamente 1 elemento, entrego el elemento
                                        #Si hay más de 1 elemento, entrego la lista de elementos
                                        #(no crea llaves al llamar a llaves no existentes)
        for v, k in invertido.items()
    }


def main_diccionario_inverso_duplicados(dict):
    print(invertir_dic_lista(dict))
````

````python
main_diccionario_inverso_duplicados(dict4)
main_diccionario_inverso_duplicados(dict5)
````

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

([1.3, 10000000000.0, 2.0, 1.0, 1.4, 1.0, 2.0, 3.0, 3.4],
['one', 'seven', '3-1/2', 'not-a-number'])

````python
lista_cliente = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

def aplanar_lista(lista):
    """Convierte una lista con sublistas/tuplas en una lista plana."""
    lista_plana = []
    for elemento in lista:
        if isinstance(elemento, (list, tuple)):
            lista_plana.extend(aplanar_lista(elemento))
        else:
            lista_plana.append(elemento)
    return lista_plana


def revisar_conver_float(obj):
    """Función que revisa si un objeto puede ser transformado a float """
    try:
        float(obj)
        return True
    except (ValueError, TypeError):
        return False


def revisar_conver_float_lista(lista):
    """Creo lista de boolean, aplicando función de revisión de convertibilidad
    input: lista de objetos
    output: lista de boolean """
    return [
        True if revisar_conver_float(elemento) else False 
        for elemento in lista
    ]

def separar_elementos(lista, lista_bool):
        """Función que crea dos listas, según convertibilidad ya evaluada
    inputs: lista objetos, lista bool convertibilidad
    output: 2 listas
    f: flag booleana, la utilizamos para dividir la lista original en dos
    """
        return (
        [e for e, f in zip(lista, lista_bool) if f],
        [e for e, f in zip(lista, lista_bool) if not f]
    )

def main_convertibilidad():
    lista_plana = aplanar_lista(lista_cliente)
    lista_bool_plana = revisar_conver_float_lista(lista_plana)
    lista_conv, lista_no_conv = separar_elementos(lista_plana, lista_bool_plana)
    imprimir_resultados("Lista de elementos convertibles a float", lista_conv,
                         "Lista de elementos no convertibles a float", lista_no_conv)
````
